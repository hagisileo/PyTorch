###    시리즈 ==> 데이터프레임 생성

```python
# 시리즈 형식
학수번호 = pd.Series(['S001','S002','S003','S004'])
교과목명 = pd.Series(['python','Excel응용','Database','컴퓨팅적사고'])
학점 = pd.Series( [3,2,3,1])
수강대상 = pd.Series(['국문','불문','영문','영문'])
```

```python
df_2 = pd.DataFrame({'학수번호':학수번호, '교과목명':교과목명, '학점':학점, '수강대상':수강대상})
df_2
```

```python
df_3 = pd.DataFrame({'교과목명':교과목명, '학점':학점, '수강대상':수강대상},
                    index = 학수번호)
df_3
```

```python
data = {
    '이름':['채치수','정대만','송태섭','서태웅','강백호','변덕규','황태산','윤대협'],
    '학교':['북산고','북산고','북산고','북산고','북산고','능남고','능남고','능남고'],
    '키':[197,184,168,187,188,202,188,190],
    '국어':[90,40,80,40,15,80,55,100],
    '영어':[85,35,75,60,20,100,65,85],
    '수학':[100,50,70,70,10,95,45,90],
    '과학':[95,55,80,75,35,85,40,95],
    '사회':[85,25,75,80,10,80,35,95],
    'sw특기':['Python','Java','Javascript','','','C','PYTHON','C#']
}
data
```

```python
#데이터 프레임 만들기
df_4 = pd.DataFrame(data)
df_4
```

```python
#이름 확인
df_4['이름'] #Name: 이름, dtype: object
df_4['키'] #Name: 키, dtype: int64
```

```python
#데이터 프레임에서 이름열 확인
df_5 = df_4[['이름', '키']] # 여러 데이터를 가져올 때 []로 한 번 더 묶음
df_5
```

```python
#새로운 데이터 프레임 생성
#data 중에서 원하는 Column만 선택하거나, 순서 변경 가능
df_new = pd.DataFrame(data, columns = ['이름', '학교', '키'])
df_new
```

## 데이터프레임 속성
*  .values   : values 속성을 확인
*  .index    : 행의 레이블 
*  .column   : 열의 레이블 관리
*  .axes     : 행과 열 레이블 관리
*  .shape    : 행,열수
*  .ndim     : 차원
*  .size     : 원소개수
*  .dtype  : 시리즈 값의 자료형
