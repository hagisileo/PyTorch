- 북산고 학생이면서 영어점수가 60점 이상인 학생

```python
#북산고 학생이면서 영어점수가 60점 이상인 학생
df.학교 == '북산고'

#영어점수가 60점 이상인 학생
df['영어'] >= 60

(df.학교 == '북산고') & (df['영어'] >= 60)
```

```python
#두가지 조건이 모두 만족하는 열을 데이터프레임으로
df[(df.학교 == '북산고') & (df['영어'] >= 60)]
```

```python
# 조건에 해당하는 학생의 수 
df[(df.학교 == '북산고') & (df['영어'] >= 60)].count()
```

```python
#조건에 해당하는 학생의 평균 영어점수는?
df[(df.학교 == '북산고') & (df['영어'] >= 60)].영어.mean()
```

- 국어와 영어 점수 중 60점 이상인 학생

```python
#데이터확인
df[(df['국어']>=60) | (df['영어']>=60)]
```

```python
# 인원수는 몇 명인가?
df[(df['국어']>=60) | (df['영어']>=60)].count()
# 키의평균은 ?
df[(df['국어']>=60) | (df['영어']>=60)].키.mean()
```
