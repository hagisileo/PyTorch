## 항목 추가 메서드
1. append : 요소 추가
2. index : 위치값을 반환 
3. insert : 리스트에 요소 삽입
4. extend : 리스트 확장

```python
food = ['피자','치킨','떡볶이','스테이크','파스타']
# 요소 추가(append)
# 감자칩 추가
food.append('감자칩')

print(food)
print()

# 요소 추가(insert)
# 카레를 3번위치 추가
food.insert(2, '카레')

print(food)
```

```python
#빈 리스트 만들고 데이터 추가하기
to_do = []

# 해야할 일을 저장하는 리스트 to_do
to_do.append('파이썬공부')

to_do
```

## 리스트 제거
1. remove : 리스트에 요소 제거
2. pop : 리스트의 맨 마지막 요소 삭제

- del 연산자 

```python
food = ['피자','치킨','떡볶이','스테이크','파스타']

# 요소 제거(del) - '떡볶이' 제거
del food[2] # ['피자','치킨','스테이크','파스타']

#마지막 항목 제거
food.pop() # '파스타' (삭제한 값)
food # ['피자','치킨','스테이크']

#리스트내 항목'치킨' 제거
food.remove('치킨') # ['피자','스테이크']
```

## 리스트 정렬
- L.reverse()  :  리스트에 있는 객체의 순서를 반대로 뒤집어 정렬
- L.sort() : 리스트의 객체를 알파벳 순서로 정렬

```python
food = ['피자','치킨','떡볶이','스테이크','파스타']

food.reverse() # ['파스타', '스테이크', '떡볶이', '치킨', '피자']

food.sort() # ['떡볶이', '스테이크', '치킨', '파스타', '피자']

food.sort(reverse=True) #['피자', '파스타', '치킨', '스테이크', '떡볶이']
print(food)
```
