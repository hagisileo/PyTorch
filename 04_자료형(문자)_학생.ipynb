# 문자열 슬라이싱(slicing)
*  문자열의 일부를 잘라보는 것.
*  문자열 인덱스를 통해서 문자열을 추출
*  슬라이싱 기호  콜론(:)
*  a[0:4] #a변수의 0번째 이상, 4번째 미만의 문자열 추출
       *  0 ≦ a  < 4  (0,1,2,3)



*   **s[0:10:2]**  
    *   s변수의 0번째 이상, 10번째 미만의 리스트를 2간격으로 추출
        
    * 0 ≦ s < 10  ('H','w','a','e',y')


[문제]결과가 나오도록 문자열 슬라이싱

```
문제해
위한
문해을위 이
문제해결을
이썬
문해을위 이
문제해결을 위한 파이썬
```



````python
과목 = '문제해결을 위한 파이썬'
print(과목[0:3]) 
print(과목[6:8]) 
print(과목[0:13:2], 과목[0:len(과목):2])
print(과목[0:5:1], 과목[:5:1], 과목[:5]) 
print(과목[10:12], 과목[10:14], 과목[10:])
print(과목[0: len(과목)])
print(과목[0: len(과목): 2])
print(과목[::2])
print(과목[:])
````
