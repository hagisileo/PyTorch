- .groupby() : 그룹화
    -  데이터프레임에서 원하는 열을 기준으로 그룹
    - df.groupby('학교') : 학교를 기준으로 묶음을 생성후 통계결과를 확인할수 있다.
    - .get_group(): 그룹화시킨 그룹정보중 선택하여 가져올 수 있다.



```python
df
```

```python
df.groupby('학교')
```

```python
# 그룹핑할 자료를 확인
# 북산고 자료를 가져오기
df.groupby('학교').get_group('북산고')
```

```python
# 그룹핑한 자료중 능남고 자료를 가져오기
df.groupby('학교').get_group('능남고')
```

```python
#그룹화된 자료의 통계값
#계산 가능한 데이터들의 평균값
df.groupby('학교').mean()
```

```python
#각 그룹의 크기
df.mean()
df.groupby('학교').size()
```

```python
#학교로 그룹화를 한 뒤에 능남고에 해당하는 데이터의 수
df.groupby('학교').size()['능남고']
```

```python
#학교로 그룹화를 한 뒤에 키의 평균 데이터
df.groupby('학교').mean()
```

```python
df.groupby('학교')['키'].mean()
```

```python
#학교로 그룹화를 한 뒤에 여러열 - 국어, 영어, 수학 평균 데이터
#학교별 국어, 영어, 수학 평균을 구하기
df.groupby('학교')[['국어','영어','수학']].mean()
```

```python
# 학년 Column 추가후 [3, 3, 2, 1, 1, 3, 2, 2] 
df['학년'] =[3, 3, 2, 1, 1, 3, 2, 2] 
```

```python
df
```

```python
# 학교별, 학년별 평균 데이터
df.groupby(['학교','학년']).mean()
```

```python
 # 학년별 평균 데이터
 df.groupby('학년').mean()
```

```python
#학년별 키열을 기준으로 정렬
df.groupby('학년').mean().sort_values('키')
```

```python
df.groupby('학년').mean().sort_values('키',ascending=False)
```

```python
#학교별 학년단위로 연산중 합계
df.groupby(['학교','학년']).sum()
```

```python
# 학교로 그룹화를 한 뒤에 각 학교별 이름과 sw특기 데이터의 수를 가져옴
df.groupby('학교')[['이름','sw특기']].count()
```

```python
# 학교로 그룹화를 한 뒤에 학년별 학생 수를 가져옴
school = df.groupby('학교')
```

```python
# 학교로 그룹화를 한 뒤에 북산고에 대해서 학년별 학생 수를 가져옴
school['학년'].value_counts()
```

```python
# 학교로 그룹화를 한 뒤에 능남고에 대해서 학년별 학생 수를 가져옴
school['학년'].value_counts().loc['능남고']
```

```python
# 학생들의 수 데이터를 퍼센트로 비교하여 가져옴
school['학년'].value_counts().loc['북산고']
```

```python
#학년별 비율
school['학년'].value_counts(normalize =True).loc['북산고']
```
