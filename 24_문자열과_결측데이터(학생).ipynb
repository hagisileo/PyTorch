```python
#국어점수가 가장 높은 학생의 이름과 학교?
df.loc[df['국어'] == max(df['국어']), ['이름', '학교']]
```

```python
# '송'씨 성을 가진 사람
df['이름'].str.startswith('송')
```

```python
#이름에 '태'가 들어가는 사람
df.loc[df['이름'].str.contains('태')]
```

```python
# 이름에 '태'가 들어가는 사람을 제외
fil = df['이름'].str.contains('태')
```

```python
df[~fil] # ~ : 반대되는 데이터
```

```python
# sw 특기가 Python이거나 Java인 사람: isin 함수
lang = ['Python', 'java']

df[df['sw특기'].isin(lang)]
```

```python
#영어 평균보다 높은 점수를 받은 학생정보
eng = df['영어'].mean()

df[df['영어'] > eng]
```

```python
df = pd.read_csv(drive_path + 'score.csv', index_col='지원번호')
df
```

- axis : index or columns
- how : any or all

```python
 # NaN row 이 하나라도 있는 삭제
df.dropna(axis = 'index', how = 'any')
```

```python
# NaN column 이 하나라도 있는 삭제
df.dropna(axis = 'columns', how= 'any')
```

```python
df['학교'] = np.nan
```

```python
# NaN column 데이터 전체가 NaN인 경우에만 column 삭제
df.dropna(axis = 'columns', how = 'all')
```
